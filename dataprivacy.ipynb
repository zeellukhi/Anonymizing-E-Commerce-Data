```python
#Tokenization for Email ID
def getEmail():
	URL_Email = "https://10minutemail.net/"
	email_page = requests.get(URL_Email)
	handle_email = BeautifulSoup(email_page.content, 'html.parser')
	email = handle_email.find(class_='div-m-0 text-c')
	email = str(email)[84:].replace("\"/></div>", '')
	return email

def clear(): 
	if os.name == 'nt':
		os.system('cls') 
	else:
		os.system('clear')

def writeToFile(email):
	file = open("emails.txt", 'a')
	write = ''
	write += email + '\n'
	file.write(write)
	file.close()

def main():
	try:
		numEmails = int(input("[+] Enter the number of emails you want to create: "))
	except:
		print("[!] Invalid Value Entered. Restarting in 2 secs...")
		time.sleep(2)
		main()
	writeOrNot = str(input("[+] Do you want to save to a file or not? (Y/N): "))
	write = False
	if 'y' == writeOrNot[0].lower():
		write = True
	if write == True:
		print("[*] Writing to file has been enabled!")
	print("[*] Creating {} emails".format(numEmails))

def random_char(y):
       return ''.join(random.choice(string.ascii_letters) for x in range(y))

def alpha_numeric(n):
    alpha = "".join(random.choice("ABCDEF") for i in range(int(n / 2)))
    res = alpha.join(random.choice("0123456789") for i in range(int(n / 2)))
    return res
```

```python
#Substitution For First 3 Characters and shuffling
df = pd.read_csv('Ecommerce Customers.csv')
for i in range (49):
    s = df.loc[i,'Email']
    new_s = s.rpartition('@')
    l=list(new_s[0])
    random.shuffle(l)
    answer = ''.join(l)
    ss = answer[0:3]
    answer.replace(ss , random_char(3))
    df.loc[i,'Email'] = answer + "@gmail.com"
    df.to_csv("Ecommerce Customers.csv", index=False)
print(df['Email'])
```

```python
#Modified Suffling of First Name + Adding Comman Random Character + Suppression (Add **) + Substitution
df = pd.read_csv('Ecommerce Customers.csv')
col = list(df)
index = col[1]
cnt=0
df=df.head(50)
s = random_char(3)
for i in df[index]:
    l = list(i)
    random.shuffle(l)
    answer = ''.join(l)
    leng_ = len(answer)
    if(leng_>=4):
        #name  = name + s
        name = answer[0:2]+random_char(3)+answer[4:leng_-2]
        name  = name + s + "**" 
        ss =  name[0:2]
        name.replace(ss , random_char(2))
        df.loc[cnt,'First Name']= name
        df.to_csv("Ecommerce Customers.csv", index=False)
    else:
        name  = name + s
        ss =  name[0:2]
        name.replace(ss , random_char(2))
        df.loc[cnt,'First Name']= name
        df.to_csv("Ecommerce Customers.csv", index=False)
    cnt+=1
df = pd.read_csv('Ecommerce Customers.csv')
print(df['First Name'])
```
